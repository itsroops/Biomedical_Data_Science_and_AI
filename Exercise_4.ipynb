**Answer 1.a)**

Confidence interval is range of values proposed that it is likely to contain the population value for an unknown parameter with certain degree of confidence (90%. 95%, 99%)

•	CI:  How accurate our estimate likes to be. How well our sample represents the population we are studying.

•	CI used when we want to express a range that we are sure that the population parameter (e.g.: Mean) lies. When we express an estimate of population parameter it is good to give it as confidence interval so we take a sample and calculate the sample mean we use it to express the range that we are sure the parameters as mean the population lies.

•	Width of confidence interval depends on two things variation of population and sample size

•	 If we have low sample variation so our estimate will be close to true population value lead to more variation, less information, wider confidence interval while larger sample will be more similar, sample error decrease, lager sample give more information, more sure about the estimate, narrow confidence intervals.

•	A 95% confidence interval is a range of values that you can be 95% certain contains the true mean of the population.
The narrower the interval (upper and lower values), the more precise is our estimate. Confidence interval is quite narrow when computed from a large sample.
As a general rule, as a sample size increases the confident interval should become more narrow.

**Answer 2.b.)**

```
# Importing the necessary package for statistical computations
from scipy import stats
import numpy as np

# Calculating sample standard deviation of all samples per gene
sd = df.iloc[:,1:39].std(axis = 1)
sqrt_of_no_of_samples = 38**0.5

# Initializing the t statistic list and p-values list for all genes
t_stat = []
p_val = []

# Computing t-statistic and p-values for all genes
for i in range(b_cap.count()):
  
  # Forming a tuple of t-statistic and gene number
  k  = (b_cap[i] - b0) / (sd[i]/sqrt_of_no_of_samples)
  tup = (i+1,k)

  # Adding the tuple to the t_statistic list
  t_stat.append(tup)

  # Forming a tuple of p-value and gene number
  k2 = stats.t.sf(np.abs(k), 38-1)*2
  tup = (i+1,k2)
  p_val.append(tup)

# Sorting the list of p-values by the second value of the tuple
p_val = sorted(p_val, key=lambda x: x[1])

# Selecting the 100 most significant genes
p_100 = p_val[:100]

# Initialize the index list for 100 significant genes
p_100_index = []

# Adding the indices for 100 significant genes
for i in p_100:
  p_100_index.append(i[0])

#print(p_100_index)
# Converting gene names dataframe to list of lists
d = df3.values.tolist()

# Initialize the gene names for 100 significant genes
p_100_names = []

# Comparing the gene index with gene names for the 100 significant genes
for i in p_100_index:
  for j in d:
    if i==j[0]:
      p_100_names.append(j[2])
print(sorted(p_100_names))
# Printing the 100 significant gene names sorted in order

#for i in p_100_names:
  #genes.append(i)
  #print(i)
```
