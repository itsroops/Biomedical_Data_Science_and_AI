```
# Importing the necessary package to read the dataset
import pandas as pd

# Reading the content of the dataset to a dataframe
df = pd.read_csv("/content/drive/My Drive/Colab Notebooks/Data Files/iris.csv")
```

```
# Printing the first five rows of the dataset
print(df.head(5))
```

```
# Importing the relevant packages for plotting
import matplotlib.pyplot as plt
import seaborn as sns

# Drawing the scatterplot of petal length against sepal length colour-coded by its species
sns.FacetGrid(df, hue="class", height = 5).map(plt.scatter, "sepal_length", "petal_length").add_legend();
plt.show()
```

The following observations can be made:
1. The group Setosa can be well spearated from the others.
2. Even Versicolor can be fairly well separated from Verginica. 
3. Hence, petal length vs sepal length is an important determiner to separate the dataset reasonably well into all the three distinct groups
4. From the point of correlation, it can be observed that for Setosa the two features are uncorrelated, for Versicolor and Verginica they have positive correlation with moderate strength.

```
# Plotting histogram for each of the variables of the dataset
df.hist(bins=30, figsize=(15, 10))
```

The following are the observations:
1. There is gap in the distribution in all the features. 
2. Sepal width is fairly normally distributed with gaps in values
3. For petal length, between the range or 3 and 7, the data is fairly normally distributed
4. For other features, not much can be inferred about the distribution pattern 

```
# Computing th correlation matrix for the features
corr_matrix = df.corr()

f, ax = plt.subplots(figsize=(80,5))

# Generating the heatmap from the correlation matrix
heatmap = sns.heatmap(corr_matrix,
                      
                      square = True,
                      linewidths = .5,
                      cmap = 'copper',
                      cbar_kws = {'shrink': .4,
                                'ticks' : [-1, -.5, 0, 0.5, 1]},
                      vmin = -1,
                      vmax = 1,
                      annot = True,
                      annot_kws = {"size": 15})


ax.set_yticklabels(corr_matrix.columns, rotation = 0)
ax.set_xticklabels(corr_matrix.columns)

sns.set_style({'xtick.bottom': True}, {'ytick.left': True})
```

The following are the observations:
1. This heatmap shows the correlation among all pairs of the feature variables
2. Petal width vs petal length has the highest positive correlation.
3. Sepal width vs petal length has the higest negative correlation. 

```
# Plotting the paiwise scatterplots for all the features
sns.pairplot(df,  hue="class")
```

Following are the observations:
1. The diagonal elements contain the data distribution of individual features
2. The number of plots is nC2, where 'n' is the number of features.
3. The upper triangular part of the plot area is symmetric to the lower trangular plot area. Hence, probing in one triangular area is sufficient
4. Setosa can be linearly separable from the other two species quite easily in all the cases. 
5. Petal length vs petal width are the most necessary features for classification into groups
6. There are minot overlaps between Versicolor and Virginia but still they can be linearly separable with minimum loss. 

```
# Plotting box plot of petal length for each species
sns.boxplot(x="class",y="petal_length",data=df)
plt.show()
```

The following are the observations:
1. The box plots are plotted for all the three classes. 
2. Box plots typically show the summary statistics like median, 25%, 75%. The whiskers further denote the extremes of the values of the plot. 
3. Setosa is again well separated from the other species using petal length.
4. Versicolour and Virginica are too well separated using petal length. However, a small number of data would overlap between the two but those data are beyond the upper quartile ranges of the two species.

```
# Plotting violin plot of sepal width for each species
sns.violinplot(x="class",y="sepal_width",data=df)
plt.show()
```

The following are the observations:
1. The violin plots shows the summary statistics like the median, lower quartile and upper quartile also shows how the data is distributed around them. 
2. Using Sepal width, we cannot separate the classes as the data would fairly overlap among one another. 

Difference between Box plots and Violin plots are as follows:
1. The violin plot combines the box plot statistics and also show the data distribution, wheareas box plot only shows the summary statistics
2. Violin plot shows the probability density of data at each values and hence more informative than box plot. 
3. A violin plot can have multiple layers whereas there is no such concept in box plot
