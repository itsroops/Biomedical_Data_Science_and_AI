**4.	What do you understand by Backpropagation and how does it work. (6)**

**Answer:**

*   
Backpropagation is a technique of training an artificial neural network for supervised learning by fine-tuning the weights and biases of a neural network on the basis of a loss function obtained in the previous epoch. 

* It computes the gradient of the loss function and transfers the error signal back from the output layer to the hidden layers that immediately contributed to the output layer.

* This process is then repeated several times for layer by layer, until each node in the network has received an error signal that describes its relative contribution to the overall error. 

* The Backpropagation algorithm calculates the minimum value of the error function using gradient descent technique.

6.	Load the breast cancer data set. Train a neural network classifier that classifies breast cancer into malignant and benign. (1000 epochs, batch_size 10, 2 hidden layers, use “adam” as optimizer) (30)


**Preparation**

Initialize tensorflow

```
# Limit GPU Usage
import tensorflow as tf
config = tf.compat.v1.ConfigProto() # Tensorflow 2.0 version
config.gpu_options.allow_growth = True
session = tf.compat.v1.Session(config=config)
```

**Import the required libraries**

```
# Importing necessary packages
import keras
from keras.models import Sequential
from keras.layers import Dense, Dropout
from keras.models import Sequential
from keras.layers import Dense
import numpy
from sklearn.metrics import confusion_matrix
import urllib
```

```
# Fixing a seed

seed = 20
numpy.random.seed(seed)
```

**Load the cancer data**

```
import pandas as pd
from sklearn.datasets import load_breast_cancer
cancer = load_breast_cancer()
```

```
# Dividing into input dataframe and output dataframe
X = pd.DataFrame(cancer.data, columns = cancer.feature_names)
Y = pd.DataFrame(cancer.target, columns = ["class"])
```

```
from sklearn.model_selection import train_test_split

# Randomly splitting the data into training and testing with a ratio of 3:1
X_train, X_test, y_train, y_test = train_test_split(X, Y, test_size=0.25, random_state = 20)
```

```
from sklearn.preprocessing import StandardScaler
#Feature Scaling of the test and train input frame

sc = StandardScaler()
X_train = sc.fit_transform(X_train)
X_test = sc.transform(X_test)
```

**a.	Evaluate the outcome by determining the loss and accuracy after training the network (10)**


```
# ANN Initialization
clf = Sequential()

# Number of input variables  = number of features = 30
# Number of nodes in first hidden layer = (30+1)/2  = 16; chosen as per best practises.

# Adding the input layer and the first hidden layer
clf.add(Dense(input_dim=30, units=16, kernel_initializer='uniform', activation='relu'))

# Adding dropout to prevent overfitting
clf.add(Dropout(rate=0.1))

# Adding the second hidden layer
clf.add(Dense(units=16, kernel_initializer='uniform', activation='relu'))
# Adding dropout to prevent overfitting
clf.add(Dropout(rate=0.1))

# Adding the output layer and using the sigmoid activtion function to obtain the output
clf.add(Dense(units=1, kernel_initializer='uniform', activation='sigmoid'))

clf.summary()
```

```
# Compiling the network developed
# Optimizer = 'adam' is chosen for gradient descent
# Binary_crossentropy is the loss function used

clf.compile(optimizer='adam', loss='binary_crossentropy', metrics=['accuracy'])
```

```
# Fitting the network
d= clf.fit(X_train, y_train, batch_size=10, epochs=1000)
```

```
# Displaying the loss for the last epoch
print("The loss for the last epoch is " + str(d.history.get('loss')[-1]))

# Displaying the accuracy for the last epoch
print("The loss for the last epoch is " + str(d.history.get('accuracy')[-1]))
```

```
# Making predictions on the test data
y_pred = clf.predict(X_test)
y_pred = (y_pred > 0.5)
```


**b.	Plot the loss and accuracy over different epochs (5)**

```
import matplotlib.pyplot as plt

# Plotting loss against epochs
plt.plot(d.history['loss'])
plt.title('Loss vs Epochs')
plt.ylabel('Loss')
plt.xlabel('Epochs')
plt.show()

# Plotting accuracy against epochs
plt.plot(d.history['accuracy'])
plt.title('Accuracy vs Epochs')
plt.ylabel('Accuracy')
plt.xlabel('Epochs')
plt.show()
```

**c.	Compute the confusion matrix and precision and recall for the predictions (10)**

```
# Printing the Confusion Matrix
from sklearn.metrics import confusion_matrix
from sklearn.metrics import accuracy_score
from sklearn.metrics import precision_score
from sklearn.metrics import recall_score

# Computing the confusion matrix
cm = confusion_matrix(y_test,y_pred)
# Displaying the confusion matrix
print("The confusion matrix is as follows:")
print(cm)
print()

# Computing the accuracy
accuracy = accuracy_score(y_test,y_pred)
# Displaying the test accuracy
print("The model's accuracy on the test set is " + str(accuracy))
print()

# Computing the precision
precision = precision_score(y_test,y_pred)
# Displaying the test accuracy
print("The model's precision on the test set is " + str(precision))
print()

# Computing the recall
recall = recall_score(y_test,y_pred)
# Displaying the test accuracy
print("The model's recall on the test set is " + str(recall))
print()
```

**7.	Evaluate the above classifier within a nested cross validation loop. (7)**

```
from sklearn.model_selection import StratifiedKFold

# define 5-fold cross validation
kfold = StratifiedKFold(n_splits=5, shuffle=True, random_state=20)
cvscores = []
for train, test in kfold.split(X, Y):


  # create model
  clf_cv = Sequential()
  # Adding the input layer and the first hidden layer
  clf_cv.add(Dense(input_dim=30, units=16, kernel_initializer='uniform', activation='relu'))

  # Adding dropout to prevent overfitting
  clf_cv.add(Dropout(rate=0.1))

  # Adding the second hidden layer
  clf_cv.add(Dense(units=16, kernel_initializer='uniform', activation='relu'))
  # Adding dropout to prevent overfitting
  clf_cv.add(Dropout(rate=0.1))
 
  # Adding the output layer and using the sigmoid activtion function to obtain the output
  clf_cv.add(Dense(units=1, kernel_initializer='uniform', activation='sigmoid'))

	# Compiling the network developed
  # Optimizer = 'adam' is chosen for gradient descent
  # Binary_crossentropy is the loss function used
  clf_cv.compile(optimizer='adam', loss='binary_crossentropy', metrics=['accuracy'])

	# Fitting the model
  clf_cv.fit(X_train, y_train, batch_size=10, epochs=1000)
 
	# evaluate the model
  scores = clf_cv.evaluate(X_test, y_test)
  print("%s: %.2f%%" % (clf_cv.metrics_names[1], scores[1]*100))
  cvscores.append(scores[1] * 100)
 
print("%.2f%% (+/- %.2f%%)" % (numpy.mean(cvscores), numpy.std(cvscores)))
```
